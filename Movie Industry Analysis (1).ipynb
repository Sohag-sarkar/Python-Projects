```python
# Import libraries

import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

plt.style.use('ggplot')
from matplotlib.pyplot import  figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(10,8)
```

```python
# Load Dataset

df=pd.read_csv("C:\\Users\\bhupe\\OneDrive\\Desktop\\Portfolio Project\\Python Project\\Movie Python\\movie.csv")
df.head()
```

```python
df.rating.value_counts().head(5)
```

```python
# Display max rows

pd.set_option('display.max_rows', None)
df.head()
```

```python
# Missing Data Percentage

for col in df.columns:
    missing_pct=np.mean(df[col].isnull())
    print('{}- {}%'.format(col, missing_pct))
```

```python
# Sorting values by Gross

sort_values= df.sort_values(by="gross", ascending= False)
sort_values.head()
```

```python
# Drop null values and store dataframe without null values in a new dataframe

new_df= df.dropna()
new_df.head()
```

```python
# Data type 

new_df.dtypes
```

```python
# Changing data types

new_df["gross"]= new_df["gross"].astype('int64')
new_df["budget"]= new_df["budget"].astype('int64')
new_df["votes"]= new_df["votes"].astype('int64')
new_df.head()
```

```python
# Drop Duplicate values

new_df["company"].drop_duplicates().sort_values(ascending = False)
new_df=new_df.drop_duplicates()
new_df.head()
```

```python
# Relationship between Budget and Gross by Scatter plot

plt.scatter(x=new_df["budget"],y=new_df["gross"])
plt.title("Budget-Gross Relationship")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
# Regression Plot

sns.regplot(x=new_df["budget"],y=new_df["gross"], scatter_kws= {"color":"blue"})
plt.show()
```

```python
# Finding Correlations

new_df.corr()
```

```python
# Correlation Heatmap

corr= new_df.corr(method='pearson')
sns.heatmap(corr)
plt.title("Correlation matrix")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```

```python
# Generating Category Codes

df_numerized= new_df
for col in df_numerized.columns:
    if (df_numerized[col].dtypes =="object"):
        df_numerized[col]= df_numerized[col].astype('category')
        df_numerized[col]= df_numerized[col].cat.codes
    
df_numerized.head()
```

```python
# Correlation Heatmap

corr= df_numerized.corr(method='pearson')
sns.heatmap(corr)
plt.title("Correlation matrix")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```

```python
# Correlations

corr_num = df_numerized.corr()
corr_num
```

```python
# Unstacking Correlations

unstack= corr_num.unstack()
unstack
```

```python
# Sorting Correlations

sort_corr= unstack.sort_values()
sort_corr
```

```python
# Finding High Correlations

high_corr= sort_corr[sort_corr> 0.5]
high_corr
```
