## 4. Average delivery perday for each delivery partner

```python
trans.head(2)
```

```python
# Group by
df5 = trans.groupby("Delivery Partner id")["Delivery attempted"]
df5
trans.sort_values("Delivery Partner id").head(5)
```

```python
# Pivot tables
df6 = pd.pivot_table(trans, index= "Delivery Partner id", values= "Delivery attempted", aggfunc= ["max","min"])
df6
```

```python
df7 = trans.groupby("Delivery Partner id")["Delivery attempted"].count()
df7
```

```python
df_merge_2 = pd.merge(df6,df7, on = "Delivery Partner id", how="left")
df_merge_2
```

```python
import numpy as npb
df_merge_2["duration"]= df_merge_2[('max', 'Delivery attempted')] - df_merge_2[('min', 'Delivery attempted')] 
df_merge_2["duration"]
df_merge_2
```

```python
# change datetime to day 
# days of working  = duration + 1 
df_merge_2["days"]= df_merge_2["duration"].dt.days + 1 
df_merge_2
```

```python
# Average delivery per day
df_merge_2["Avg_delivery"]= df_merge_2["Delivery attempted"]/df_merge_2["days"]
df_merge_2[ "Avg_delivery"]
```

```python
# Save the data to excel
df8 = df_merge_2[ "Avg_delivery"]
df8.to_excel("C:\\Users\\bhupe\\OneDrive\\Desktop\\Portfolio Project\\avg_deli.xlsx")
```

## 6. Average TAT for each assignee


```python
# Find max delivery date according to id
df10 = deli.groupby("ID").max()
df10
```

```python
# Merge trans and max delivery data
df_merge_4 = pd.merge(trans, df10, left_on = "Id", right_on= "ID", how= "left")
df_merge_4.head(15).sort_values("Id")
```

```python
# Create new dataframe according to columns
df11= df_merge_4[["Asignee name","Delivery date","order date"]]
df11.head(5)
```

```python
# find delivery duration
df11["duration"]= df11["Delivery date"] - df11["order date"] 
df11["days"]= df11["duration"].dt.days
df11.head(15)
```

```python
# Find TAT per Assignee
df11.groupby("Asignee name").mean()
```
